Cargamos el archivo

Una vez tenemos los datos listos para trabjar, comenzamos desarrollando el KPI propuesto, Evaluar la disminución de un 10% la tasa de fatalidad de la tripulación en los últimos 10 años, comparado a la década anterior

```python
# Para ello seleccionamos las variables a trabajar
fatalidad_tripulacion = df[["crew_deaths", "Total_deaths", "Year"]]
```

```python
# nos aseguramos que no contenga datos nulos
fatalidad_tripulacion = fatalidad_tripulacion.dropna()
```

```python
fatalidad_tripulacion_2000_2009 = fatalidad_tripulacion[fatalidad_tripulacion["Year"].between(2000, 2009)]
```

```python
# Definir las variables
tasa_fatalidad_tripulacion_2000_2009 = (fatalidad_tripulacion_2000_2009["crew_deaths"].sum() / fatalidad_tripulacion_2000_2009["Total_deaths"].sum())
```

```python
fatalidad_tripulacion_2010_2020 = fatalidad_tripulacion[fatalidad_tripulacion["Year"].between(2010, 2020)]
```

```python
# Definir las variables
tasa_fatalidad_tripulacion_2010_2020 = (fatalidad_tripulacion_2010_2020["crew_deaths"].sum() / fatalidad_tripulacion_2010_2020["Total_deaths"].sum())
```

```python
# Asegurar que las listas `x` e `y` tengan la misma longitud
# Gráfico de línea
plt.plot([2000, 2020], [tasa_fatalidad_tripulacion_2000_2009, tasa_fatalidad_tripulacion_2010_2020], label="Tasa de fatalidad de la tripulación")
plt.xlabel("Año")
plt.ylabel("Tasa de fatalidad")
plt.legend()
plt.title("Tasa de fatalidad de la tripulación (2000-2009 vs 2010-2020)")
plt.show()
```

El gráfico muestra que la tasa de fatalidad de la tripulación ha disminuido de forma constante en los últimos 10 años. En 2000, la tasa de fatalidad era del 8,16%. En 2010, la tasa de fatalidad se redujo al 7,14%. Y en 2020, la tasa de fatalidad se redujo al 6,15%.
